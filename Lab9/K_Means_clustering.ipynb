```python
from sklearn.datasets import make_blobs
X, _ = make_blobs(n_samples=100, centers=3, n_features=2, cluster_std=0.5, random_state=0)
```

```python
print(X)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(X[:,0], X[:,-1])
plt.xlabel('Feature1')
plt.ylabel('Feature2')
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans= KMeans(n_clusters=3, random_state=121).fit(X)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
plt.scatter(X[:,0],X[:,-1])
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], c='red', marker='x')
plt.title('Data points and cluster centroids produced by Kmeans')
plt.show()
```

```python
#silhouette_score
from sklearn.metrics import silhouette_score
print(silhouette_score(X,kmeans.labels_))
```

```python
# Import the KElbowVisualizer method
from yellowbrick.cluster import KElbowVisualizer
# Instantiate a scikit-learn K-Means model
model = KMeans(random_state=122)
# Instantiate the KElbowVisualizer with the number of clusters and the metric
visualizer = KElbowVisualizer(model, k=(2,6), metric='silhouette',timings=False)
# Fit the data and visualize
visualizer.fit(X)
visualizer.poof()
```
