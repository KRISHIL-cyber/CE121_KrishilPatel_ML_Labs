```python
# Sigmoid function
def sigmoid(z):
    return 1/(1 + np.exp(-z))
```

```python
#Loss function
def loss_func(y, a):
    total_bce_loss = np.sum(-a * np.log(y) - (1 - a) * np.log(1 - y))
    m = y.shape[0]
    j = total_bce_loss /m
    return j
```

```python
dw = []
db = 0
J = 0
alpha = 0.1
for x in range(len_x):
    dw.append(0)
```

```python
def predict(input):
    z = np.dot(input, w) + b
    h= sigmoid(z)
    for i in range(len(h)):
        if(h[i]>=0.5):
            h[i]=1
        else:
            h[i]=0
    return h
```

```python
#Repeating the process 3000 times
for i in range(3000):
    z = np.dot(X_train, w) + b
    y_pred = sigmoid(z)
    l = loss_func(y_pred, y_train)
    dw = np.dot((y_pred-y_train).T, X_train)/X_train.shape[0]
    db = np.mean(y_pred-y_train)
    w = w - alpha * dw
    b = b - alpha* db
    print("Round:",i,"Weight:",w,"Bias:",b)
```

```python
import numpy as np
import pandas as pd
import io
import matplotlib.pyplot as plt
import torch.nn as nn
import torch
```

```python
data = pd.read_csv('/content/drive/MyDrive/Datasets/BuyComputer.csv')

data.drop(columns=['User ID',],axis=1,inplace=True)
data.head()
```

```python
Y=data.iloc[:,-1].values
X=data.iloc[:,:-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y, random_state=121)
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression(random_state = 121)
```

```python
from sklearn.metrics import  accuracy_score
#Fit
LR.fit(X_train,y_train)

y_pred = LR.predict(X_test)
test_accuracy = accuracy_score(y_test, y_pred)
print('Accuracy: ', test_accuracy)
```
